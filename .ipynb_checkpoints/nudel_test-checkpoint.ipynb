# Level scheme plots

```python
def level_scheme(nuc=None, nucleons=None, protons=None, n_max = 40):
    decays = []
    if not nuc:
        nuc = Nuclide(nucleons, protons)
    for level in nuc.adopted_levels.levels[:n_max]:
        decays.extend(level.decays)
    
    i = 0.5
    for level in nuc.adopted_levels.levels[:n_max]:
        plt.axhline(level.energy.val, color='k')
        for decay in level.decays:
            plt.plot(
                [i, i],
                [decay.orig_level.energy.val, decay.dest_level.energy.val],
            )
            i += 1
    plt.xlim(0, i - 0.5)
    plt.ylim(0, nuc.adopted_levels.levels[n_max].energy.val)
    plt.show()
    
# level_scheme(nucleons=20, protons=10, n_max=10)
```

# Load transition from ENSDF

```python
class Gamma():
    """
    container class for Gamma (M1,E2) transitions
    """
    def __init__(self,A,Z,N,multipolarity):
        self.multipolarity = multipolarity
        self.A = A
        self.Z = Z
        self.N = N
        
        self.twoJi = None
        self.twoJf = None
        self.ni = None
        self.nf = None
        
        self.Ei = None
        self.Ef = None
        self.dEi = None
        self.dEf = None
        
        self.BW = None
        self.dBW = None
        
#         self.Thalf = None
        self.ref = None
        
        
    def fill(decay,twoJi,twoJf,ni,nf):
        self.twoJi = twoJi
        self.twoJf = twoJf
        self.ni = ni
        self.nf = nf
        
        self.Ei = decay.orig_level.energy.val
        self.Ef = decay.dest_level.energy.val
        self.dEi = decay.orig_level.energy.pm
        self.dEf = decay.dest_level.energy.pm
            
        o_string = f'B{multipolarity}W'
        self.BW = decay.attr[o_string].val
        self.dBW = decay.attr[o_string].pm
        
#         self.Thalf = None
        self.xref = decay.xref
        
```

```python
def clean_levels(levels):
    keep = []
    for l in levels:
        if not len(l.ang_mom)>0:
            continue
        if not l.ang_mom[0].parity=='+':
            continue
        keep.append(l)
    return keep
```

```python
levels = nuc.adopted_levels.levels[:n_levels]
levels = clean_levels(levels)
```

```python
def parse_ang_mom_twoj(x):
    if len(x)==1:
        x = x[0]
    else:
        raise ValueError('Indeterminate ang mom entry: ' + str(x))
    
    p = str(x.parity)
    twoj = int(2*x.val)

#     if len(x)==2 and ( x[-1] in ['+','-'] ):
#         return [2*int(x[0]), x[-1]]
#     if len(x)==4 and ( x[-1] in ['+','-'] ):
#         return [int(x[0]), x[-1]]
#     else:
#         raise ValueError('Invalid ang mom entry: ' + str(x))
        
    return [twoj, p]

def get_J_list_from_levels(levels):
    return [parse_ang_mom_twoj(l.ang_mom) for l in levels]

def times_so_far(ls):
    out = [0]*len(ls)
    for i in range(len(ls)):
        out[i] = ls[:i+1].count(ls[i])
    return out

def append_count(jp_list):
    counts = times_so_far(jp_list)
#     ic(counts)
    for i,x in enumerate(jp_list):
        x.append(counts[i])
    return jp_list

jp_list = get_J_list_from_levels(levels)
append_count(jp_list)
```

```python
def collect_transitions(transition_type,nuc=None, nucleons=None, protons=None, n_max = 40):
    decays = []
    if not nuc:
        nuc = Nuclide(nucleons, protons)
    for level in nuc.adopted_levels.levels[:n_max]:
        decays.extend(level.decays)
    
    trans_list = []
    for decay in decays:
        print(decay.attr)
        if transition_type in decay.attr.keys():
            trans_list.append(decay)
#             print('Added to the list!')
            
    return trans_list
```
